## 3. Display the percentage distribution of the data in each year using the pie chart.

```python
# Storing the data of total production of honey done in each year also string all unique years in label variable  

pie_data = honey_p.groupby('year')['totalprod'].sum()
label = honey_p['year'].unique()
```

```python
# plotting the pie chart
plt.figure(figsize=(8,8))
plt.pie(pie_data,labels= label,autopct='%1.1f%%',startangle = 90,counterclock=False)
plt.show()
```

Inference:
1. From 1998 - 2000 the production of honey was either more than 8% or very close to 8%
2. From 2000 onwards the most of production are below 7%   

```python
# Storing the data of total production value of honey done in each year also string all unique years in label1 variable  


pie_data2 = honey_p.groupby('year')['prodvalue'].sum()
label1 = honey_p['year'].unique()
```

```python
plt.figure(figsize=(8,8))
plt.pie(pie_data2,labels=label1,autopct = '%1.1f%%',startangle = 90,counterclock = False)
plt.show()
```

Inference:
1. From 1998 - 2007 price percentage seems to be fare expect for year 2002, 2003 & 2004
2. We can see some unusal price hike between 2002 - 2004
3. After 2007 the price increased by almost 2% and gradually it kept on increasing.

**From above two pie chart we can say that as there is a decline in production there is increase in price**

## 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
# Plotting Box Plot between year and prodvalue

plt.figure(figsize=(8,5))
plt.title("Box Plot: Year & Production Value")
plt.xlabel('Year')
plt.ylabel('Production Value')
sns.boxplot(data = honey_p,x='year',y='prodvalue')
plt.grid()
plt.show()
```

Inference:
1. Porduction value for all year are mostly between 0.1 to 1.5 after approx 1.5 there are only outliers.
2. Most of the outliers lies between 1 and 3 dollars. 
3. For last 5 Years (2008-2012) seems to consistently have oultliers above 4. 
4. In year (2010) it reaches 7 which very far from max limit of year 2010.
